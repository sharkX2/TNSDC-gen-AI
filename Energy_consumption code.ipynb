```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Energy_consumption.csv')

# Assuming 'column_name' is the column you want to summarize
column_name = 'EnergyConsumption'

# Calculate the counts of unique values in the specified column
value_counts = data[column_name].value_counts()

# Plot a bar graph with black and blue bars
plt.figure(figsize=(10, 6))
value_counts.plot(kind='bar', color=['black', 'blue'])  # Change color to black and blue
plt.title('Summary of {}'.format(column_name))
plt.xlabel('{}'.format(column_name))
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
